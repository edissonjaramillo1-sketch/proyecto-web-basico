```python

# Ingreso y validación del primer conjunto

import itertools  # Librería necesaria para combinaciones y productos más adelante

# Variable que indica si el conjunto es válido
conjunto_valido = True

# Pedir al usuario que ingrese los elementos del conjunto separados por coma
entrada = input("Ingrese los elementos del conjunto separados por coma (Enter para vacío): ")

# Si el usuario no escribe nada, el conjunto es vacío
if entrada.strip() == "":
    conjunto = []
else:
    # Convertir la entrada en lista separando por comas
    conjunto = [x.strip() for x in entrada.split(",")]

    # Validar duplicados
    if len(conjunto) != len(set(conjunto)):
        print("Error: los conjuntos no pueden tener elementos repetidos")
        conjunto_valido = False

    # Validar tipo uniforme
    elif all(x.isdigit() for x in conjunto):
        conjunto = [int(x) for x in conjunto]  # convertir a enteros
        conjunto_valido = True
    elif all(x.isalpha() for x in conjunto):
        conjunto_valido = True
    else:
        print("Error: todos los elementos del conjunto deben ser del mismo tipo (números o letras)")
        conjunto_valido = False

# Mostrar el conjunto solo si es válido
if conjunto_valido:
    print("Conjunto ingresado:", conjunto)
else:
    print("No se puede usar este conjunto.")
```

```python
# subconjuntos

# Verificar primero si el conjunto es válido
if conjunto_valido:

    # Crear una lista donde se almacenarán todos los subconjuntos
    subconjuntos = []

    # Generar subconjuntos de todos los tamaños posibles
    for i in range(len(conjunto)+1):

        # Generar combinaciones de tamaño i usando itertools
        for comb in itertools.combinations(conjunto, i):
            subconjuntos.append(comb)

    # Mostrar todos los subconjuntos generados
    print("Subconjuntos:")
    for s in subconjuntos:
        print(s)

    # Mostrar cuántos subconjuntos existen
    print("Cantidad de subconjuntos:", len(subconjuntos))

else:
    print("No se puede calcular debido a elementos duplicados.")
```

```python
# subconjuntos propios

# Verificar primero si el conjunto es válido
if conjunto_valido:

    # Crear una lista para almacenar los subconjuntos propios
    subconjuntos_propios = []

    # Un subconjunto propio es cualquier subconjunto diferente al conjunto original
    for s in subconjuntos:
        if set(s) != set(conjunto):
            subconjuntos_propios.append(s)

    # Mostrar los subconjuntos propios
    print("Subconjuntos propios:")
    for s in subconjuntos_propios:
        print(s)

    # Mostrar la cantidad de subconjuntos propios
    print("Cantidad de subconjuntos propios:", len(subconjuntos_propios))

else:
    print("No se puede calcular debido a elementos duplicados.")
```

```python

# Particiones de un conjunto

import itertools

# Verificamos si el conjunto es válido antes de generar particiones
if conjunto_valido:

    # Función recursiva para generar todas las particiones de un conjunto
    def generar_particiones(conjunto):
        if not conjunto:
            return [[]]  # caso base: conjunto vacío -> una partición vacía
        else:
            primera = conjunto[0]  # tomamos el primer elemento
            resto = conjunto[1:]   # el resto del conjunto
            particiones_resto = generar_particiones(resto)  # particiones del resto
            resultado = []
            # Recorremos todas las particiones ya generadas del resto
            for particion in particiones_resto:
                # Opción 1: poner el elemento en un grupo nuevo
                resultado.append([[primera]] + particion)
                # Opción 2: añadir el elemento a cada grupo existente
                for i, grupo in enumerate(particion):
                    nueva = [g for g in particion]  # copiamos la partición
                    nueva[i] = [primera] + grupo    # agregamos el elemento al grupo i
                    resultado.append(nueva)
            return resultado

    # Generamos las particiones usando la función
    particiones = generar_particiones(conjunto)

    # Mostrar todas las particiones
    print("Particiones del conjunto:")
    for p in particiones:
        print(p)

    # Mostrar el número de particiones (Número de Bell)
    print("Número de particiones (Número de Bell):", len(particiones))

else:
    print("No se puede calcular debido a elementos duplicados.")
```

```python

# Conjunto Potencia

import itertools

# Verificamos si el conjunto es válido antes de calcular el conjunto potencia
if conjunto_valido:

    # El conjunto potencia incluye todos los subconjuntos posibles, incluidos el vacío y el conjunto completo
    conjunto_potencia = []
    for i in range(len(conjunto)+1):
        for comb in itertools.combinations(conjunto, i):
            conjunto_potencia.append(list(comb))  # convertimos la tupla en lista para mostrar

    # Mostrar conjunto potencia
    print("Conjunto potencia:")
    for s in conjunto_potencia:
        print(s)

    # Mostrar la cantidad de elementos en el conjunto potencia
    print("Número de elementos en el conjunto potencia:", len(conjunto_potencia))

else:
    print("No se puede calcular debido a elementos duplicados.")
```
